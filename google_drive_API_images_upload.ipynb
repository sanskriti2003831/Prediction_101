```python
import os
from google.oauth2.service_account import Credentials
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
from googleapiclient.errors import HttpError
```

```python
#Uploading Runway Images

# Define the required scope
SCOPES = ['https://www.googleapis.com/auth/drive']

# Authenticate using the service account JSON key
creds = Credentials.from_service_account_file(
    "/Users/user/Desktop/Prediction_101/drive-api-454402-a7abaf2ff3d9.json",
    scopes=SCOPES
)

# Initialize the Drive API service
drive_service = build('drive', 'v3', credentials=creds)
print("Google Drive API connected successfully!")

# Set the path to the folder containing the PNG files
folder_path = "/Users/user/Desktop/Project Submission/Implementation/Source Code/Source Code/project 2/public/images/AllRunwayImages"

# Set the Google Drive folder ID
folder_id = '1ooFANAmsEZiGvvUXObDg-WvmdfnNKbmQ'

# Check if the folder exists on Google Drive
try:
    folder = drive_service.files().get(fileId=folder_id).execute()
    print(f"Folder exists on Drive: {folder['name']}")
except HttpError as error:
    print(f"Error: Folder not found - {error}")
    exit()

# Check if the local folder exists
if not os.path.exists(folder_path):
    print(f"Error: The local folder '{folder_path}' does not exist.")
else:
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)

        # Only upload .png files
        if os.path.isdir(file_path) or not filename.lower().endswith('.png'):
            continue

        # Upload the file
        file_metadata = {
            "name": filename,
            "parents": [folder_id]
        }
        media = MediaFileUpload(file_path, mimetype="image/png")
        file = drive_service.files().create(
            body=file_metadata,
            media_body=media,
            fields="id"
        ).execute()

        print(f"Uploaded: {filename} (File ID: {file.get('id')})")
```
